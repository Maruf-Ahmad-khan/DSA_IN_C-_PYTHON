```python
# CLT (Central Limint Theorem)
# Set the parameters
num_samples = 1000
sample_size = 30
distribution_range = (0, 1)
# Generate samples from a uniform distribution
sample = np.random.uniform(distribution_range[0], 
                           distribution_range[1],
                           (num_samples, sample_size))
# calculate the sample means
sample_means = np.mean(sample, axis= 1)
# plot the histogram of the sample mean
plt.hist(sample_means, bins = 30, density= True, edgecolor = 'black')
plt.title("Histogram of sample Means")
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.show()
```

```python
# Exponential distribution
num_samplesd = 1000
sample_sized = 50
lambda_param = 2
samples = np.random.exponential(scale=1/lambda_param, size=(num_samplesd, sample_sized))
sample_mean = np.mean(samples , axis=1)
# plot the histogram of the sample mean
plt.hist(sample_means, bins = 30, density= True, edgecolor = 'black')
plt.title("Histogram of sample Means (Exponential Distribution)")
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_sample_distribution(population_mean, population_std, sample_size, num_samples):
    sample_means = []
    
    for _ in range(num_samples):
        # Generate a random sample from a normal distribution with the given mean and standard deviation
        sample = np.random.normal(population_mean, population_std, sample_size)
        
        # Calculate the mean of the sample and add it to the list
        sample_mean = np.mean(sample)
        sample_means.append(sample_mean)
    
    return sample_means

# Population parameters
population_mean = 50
population_std = 10

# Sample parameters
sample_size = 30
num_samples = 1000

# Generate the sample distribution
sample_distribution = generate_sample_distribution(population_mean, population_std, sample_size, num_samples)

# Plot the sample distribution
plt.hist(sample_distribution, bins=30, edgecolor='black', alpha=0.7)
plt.title('Sample Distribution of the Mean')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.show()
```
